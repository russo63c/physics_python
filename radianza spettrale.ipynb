```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const

# Planck's law for blackbody radiation as a function of wavelength
def planck_wavelength(wavelength, T):
  """Calculates the spectral radiance of a blackbody at a given wavelength and temperature.

  Args:
    wavelength: Wavelength in meters.
    T: Temperature in Kelvin.

  Returns:
    Spectral radiance in W/(m^3 sr).
  """
  h = const.h  # Planck's constant
  c = const.c  # Speed of light
  k = const.k  # Boltzmann constant

  # Avoid division by zero or very small numbers for numerical stability
  if wavelength <= 0 or T <= 0:
      return 0

  # Planck's law
  radiance = (2 * h * c**2 / wavelength**5) / (np.exp((h * c) / (wavelength * k * T)) - 1)
  return radiance

# Set the temperature (in Kelvin)
temperature = 3000  # Example temperature

# Set the range of wavelengths (in meters)
wavelengths = np.linspace(1e-9, 3e-6, 500)  # From 1 nm to 3 micrometers

# Calculate the spectral radiance for each wavelength
radiance_values = [planck_wavelength(w, temperature) for w in wavelengths]

# Plot the blackbody spectrum
plt.figure(figsize=(10, 6))
plt.plot(wavelengths * 1e9, radiance_values)  # Convert wavelength to nm for plotting
plt.xlabel("Lunghezza d'onda (nm)")
plt.ylabel("Radianza spettrale (W/(m$^3$ sr))")
plt.title(f"Spettro di corpo nero a {temperature} K")
plt.grid(True)
plt.show()
```

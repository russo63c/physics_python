**Reticolo di diffrazione**

```python
import numpy as np
import matplotlib.pyplot as plt

L= 3.0 #distanza in metri dello schermo dalla fenditura
a=0.5e-4 # apertura fenditura in m
d=1.e-4 # distanza tra le fenditure
N = 10000000 # Numero di fenditure

l=650e-9 # lunghezza d'onda laser rosso
#l=532e-9 # lunghezza d'onda laser verde
#l=405e-9 # lunghezza d'onda laser viola

I_max=1000
```

```python
# Define the x values
x = np.linspace(-1, 1, 1000)

beta=np.pi*a*np.sin(x/L)/l
gamma=np.pi*d*np.sin(x/L)/l

# Compute the I values
I = I_max*((np.sin(beta)/beta)**2)*((np.sin(N*gamma)/np.sin(gamma))**2)

# Create the plot
plt.figure(figsize=(10, 6))

# Plotting the sine wave
plt.plot(x, I, label=r'red laser', color='red')


plt.axis([-1, 1, 0, 1000])
# Adding titles and labels
plt.title('Fraunhofer diffraction')

plt.xlabel('x [m]')
plt.ylabel('I value')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Display the plot
plt.show()
```

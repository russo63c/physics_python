```python
import numpy as np
import matplotlib.pyplot as plt

input_file = "Dati pianeta.txt"  # Nome del file di input
f=open(input_file, 'r')
G=6.67e-11    #Costante di gravitazione Universale unità di misura SI

riga = f.readline()
pianeta=riga #Pianeta
riga = f.readline()
raggio=float(riga) #raggio del Pianeta m
riga = f.readline()
massa=float(riga)  #massa del Pianeta kg

g=G*massa/raggio**2
print("l'accelerazione di gravità sul pianeta", pianeta," è: ",g)

hmax=float(input("Inserisci l'altezza di caduta del corpo (m) "))
t = np.linspace(0, 100, 1000)
h=hmax-0.5*g*t**2

# Create the plot
plt.figure(figsize=(10, 6))


plt.plot(t, h, label=r'$h = h0-0,5gt^2$', color='red')

# Adding titles and labels
titolo='Caduta di un grave con accelerazione di gravità g='+str(round(g,2))
plt.title(titolo)
plt.axis([0, 100, 0, hmax*1.1])
plt.xlabel('Tempo(s)')
plt.ylabel('Altezza (m)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Display the plot
plt.show()
```

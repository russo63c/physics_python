```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Parametri del moto circolare uniformemente accelerato
r = 3 # Raggio della circonferenza
alpha = 0.1 # Accelerazione angolare (rad/s2)
omega0 = 0 # Velocit√† angolare iniziale (rad/s)
theta0 = 0 # Posizione angolare iniziale (rad)

# Calcolo posizione angolare in funzione del tempo: theta(t) = theta0 + omega0*t + 0.5*alpha*t**2

def posizione_angolare(t):
  return theta0 + omega0 * t + 0.5 * alpha * t**2
```

```python
# Setup dell'animazione
fig, ax = plt.subplots(figsize=(5, 5)) # Figura quadrata per la circonferenza
ax.set_xlim(-r - 1, r + 1) # Limiti dell'asse x
ax.set_ylim(-r - 1, r + 1) # Limiti dell'asse y
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Moto circolare uniformemente accelerato")
ax.set_aspect('equal') # Assicura che la circonferenza sia un cerchio

# Creazione della sfera
sfera, = ax.plot([], [], 'o', color='blue', markersize=12)


# Creazione della circonferenza
cerchio = plt.Circle((0, 0), r, color='black', fill=False) # Centro in (0, 0)
centro = plt.Circle((0, 0), 0.1, color='black', fill=True) # Centro in (0, 0)
cerchiofix= plt.Circle((3.5, 0), 0.1, color='black', fill=True) # Centro in (0, 0)
ax.add_artist(cerchio)
ax.add_artist(centro)
ax.add_artist(cerchiofix)


```

```python
# Inizializzazione dell'animazione
def init():
  sfera.set_data([], [])
  return sfera,

# Funzione di aggiornamento per ogni fotogramma
def update(frame):
  t = frame / 25.0 # Tempo in secondi (frame-rate di 10 fps)
  theta = posizione_angolare(t) # Calcolo della posizione angolare theta(t)
  x = r * np.cos(theta) # Calcolo della coordinata x
  y = r * np.sin(theta) # Calcolo della coordinata y
  sfera.set_data([x], [y]) # Aggiorna la posizione della sfera
  return sfera,

# Creazione dell'animazione (20 secondi a 10 fps = 200 frames)
frames = 400
ani = FuncAnimation(fig, update, frames=frames, init_func=init, blit=True, interval=40)

# Mostra l'animazione in formato HTML5

display( HTML(ani.to_jshtml()))
```

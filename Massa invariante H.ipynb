
**Massa del Bosone di Higgs**

Nei dati sono presenti tre eventi registrati con il CMS nei quali sono stati identificati 4 muoni simultaneamente.

i tre eventi sono candidati ad essere identificati come prodotti del decadimento dell'Higgs.

Per ogni evento è stata misurata l'Energia totale e le tre componenti della quantità di moto.

Le unità di misura delle grandezze misurate è $GeV/c^2$ per l'energia e $GeV/c$
per la quantità di modo, per entrambi è posto $c=1$.


La massa invariante di ciascun muone è $m_\mu = \sqrt(E^2-\vec p^2c^2)$

La massa invariate della particella madre è data da:


$$ E_{tot}^2 = (E_1+E_2+E_3+E_4)^2$$

$$ P_{tot}^2 = (p_{1x}+p_{2x}+p_{3x}+p_{4x})^2+(p_{1y}+p_{2y}+p_{3y}+p_{4y})^2(p_{1z}+p_{2z}+p_{3z}+p_{4z})^2$$

e infine:

$$M_H = \sqrt(E_{tot}^2-P_{tot}^2c^2)$$

La massa a riposo dei muoni è $m_\mu = 0,1057 \ GeV/c^2$

mentre la massa del Bosone di Higgs così determinata è

$$m_H = 123,9 GeV/c^2$$

Nei dati sono presenti quattro eventi registrati con il CMS nei quali sono stati identificati 4 muoni simultaneamente.

i 4 eventi sono candidati ad essere identificati come prodotti del decadimento dell'Higgs.

Per ogni evento è stata misurata l'Energoa totale e le tre componenti della quantità di moto.

Le unità di misura delle grandezze misurate è $MeV/c^2$ per l'energia e $MeV/c$
per la quantità di modo, per entrambi è posto $c=1$.


La massa invariante di ciascun muone è $m_\mu = \sqrt(E^2-\vec p^2)$

La massa invariate della particella madre è data da:


$$ E^2 = (E_1+E_2+E_3+E_4)^2$$

$$ P_{tot}^2 = (p_{1x}+p_{2x}+p_{3x}+p_{1x})^2+$$

```python
import numpy as np

# Primo Evento dati E, px,py,pz
l1=np.array([129.372846913,22.9629961379,-9.926534504,126.931032718])
l2=np.array([83.5239758757,-27.8565549044,12.7925529551,77.6956003345])
l3=np.array([16.4489924484,-13.1483136251,8.50097813356,5.04117046605])
l4=np.array([9.84524507523,2.62316570872,-8.10533477669,-4.93358165269])

# Secondo Evento dati E, px,py,pz
#l1=np.array([27.1459627834,19.4504684626,-4.77787612605,-18.3236042167])
#l2=np.array([41.4548617806,-9.97197945808,22.908837764,-33.0794549982])
#l3=np.array([15.7785249216,12.307407592,5.03077282075,8.49593382931])
#l4=np.array([103.2195471,-10.5611261152,-26.435982513,-99.2163108915])

# terzo Evento dati E, px,py,pz
#l1=np.array([92.5961775474,8.5353921252,-22.5757527987,-89.3953208549])
#l2=np.array([59.8124628499,-10.7217014151,41.8109883787,-41.4053850271])
#l3=np.array([21.4101492687,6.95334165376,-20.2443480232,0.460330755882])
#l4=np.array([11.0130223689,-7.74692375716,6.87465686858,-3.74311724054])

#Calcola massa invariante muoni
m1=l1[0]**2-(l1[1]**2+l1[2]**2+l1[3]**2)
print(m1**0.5)
m2=l2[0]**2-(l2[1]**2+l2[2]**2+l2[3]**2)
print(m2**0.5)
m3=l3[0]**2-(l3[1]**2+l3[2]**2+l3[3]**2)
print(m3**0.5)
m4=l4[0]**2-(l4[1]**2+l4[2]**2+l4[3]**2)
print(m4**0.5)

# Calcola Massa Invariante Higgs
p=np.array([1,1,1])
p[0]=l1[1]+l2[1]+l3[1]+l4[1]
p[1]=l1[2]+l2[2]+l3[2]+l4[2]
p[2]=l1[3]+l2[3]+l3[3]+l4[3]
ptot2=p[0]**2+p[1]**2+p[2]**2
E2=(l1[0]+l2[0]+l3[0]+l4[0])**2
MassH=E2-ptot2
print((MassH)**0.5)
```

**Confronto con la massa attesa del Bosone di Higgs**

La massa attesa del Bosone di Higgs è circa $125 \ GeV/c^2$.

I valori della massa invariante calcolati per i tre eventi sono:

*   Evento 1: 122.67 GeV/c²
*   Evento 2: 121.89 GeV/c²
*   Evento 3: 127.05 GeV/c²

Questi valori sono relativamente vicini alla massa attesa, il che supporta l'ipotesi che questi eventi possano essere decadimenti del Bosone di Higgs.

```python
import numpy as np

# Dati degli eventi (E, px, py, pz) per 3 eventi, ciascuno con 4 muoni
events_data = [
    # Primo Evento (4 muoni)
    [np.array([129.372846913, 22.9629961379, -9.926534504, 126.931032718]),
     np.array([83.5239758757, -27.8565549044, 12.7925529551, 77.6956003345]),
     np.array([16.4489924484, -13.1483136251, 8.50097813356, 5.04117046605]),
     np.array([9.84524507523, 2.62316570872, -8.10533477669, -4.93358165269])],

    # Secondo Evento (4 muoni)
    [np.array([27.1459627834, 19.4504684626, -4.77787612605, -18.3236042167]),
     np.array([41.4548617806, -9.97197945808, 22.908837764, -33.0794549982]),
     np.array([15.7785249216, 12.307407592, 5.03077282075, 8.49593382931]),
     np.array([103.2195471, -10.5611261152, -26.435982513, -99.2163108915])],

    # Terzo Evento (4 muoni)
    [np.array([92.5961775474, 8.5353921252, -22.5757527987, -89.3953208549]),
     np.array([59.8124628499, -10.7217014151, 41.8109883787, -41.4053850271]),
     np.array([21.4101492687, 6.95334165376, -20.2443480232, 0.460330755882]),
     np.array([11.0130223689, -7.74692375716, 6.87465686858, -3.74311724054])]
]

# Calcola massa invariante per ciascun evento
for i, event in enumerate(events_data):
    print(f"\nEvento {i+1}:")
    total_E = 0
    total_px = 0
    total_py = 0
    total_pz = 0

    for muon in event:
        total_E += muon[0]
        total_px += muon[1]
        total_py += muon[2]
        total_pz += muon[3]

    # Calcola Massa Invariante Higgs
    ptot2 = total_px**2 + total_py**2 + total_pz**2
    E2 = total_E**2
    MassH = E2 - ptot2
    print(f"  Massa Invariante Higgs: {MassH**0.5}")
```

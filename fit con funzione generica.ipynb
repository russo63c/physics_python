```python
#@title Fit con funzione generica
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

xdata = [0.0,0.34,0.69,1.03,1.38,1.72,2.07,2.41,2.76,3.1,3.45,3.79,4.14,4.48,4.83,5.17,5.52,5.86,6.21,6.55,6.9,7.24,7.59,7.93,8.28,8.62,8.97,9.31,9.66,10.0]
ydata = [8.55,8.38,11.83,10.82,5.23,10.73,10.0,10.38,12.33,18.52,25.9,30.15,28.73,31.86,41.57,50.64,57.24,57.95,67.84,73.98,78.8,96.7,98.41,111.97,118.44,128.5,140.57,154.77,161.76,181.47,]

xdata = np.asarray(xdata)      #trasformazione in array di numpy
ydata = np.asarray(ydata)

def myfun(x,A,B,C,D):            #definizione della funzione di fit
  return A*x**3 + B*x**2 + C*x+D

par,cov = curve_fit(myfun,xdata,ydata)      #fit ai dati e inserimento dei valori nelle liste par e cov
for i in range(len(par)):
  print('parametro ',i,'=',round(par[i],3)) #visualizziamo i valori dei parametri
  print('errore parametro ',i,'=',round(cov[i][i],3)) #visualizziamo i valori degli errori sui parametri
  print()

xfit = np.linspace(-1,10,200)               #per disegnare la curva, servono altre due liste (x,y)
yfit = myfun(xfit,par[0],par[1],par[2],par[3])     #la y si calcola come y=f(x) dove f(x) Ã¨ proprio la parabola di fit. par[0] si intenda il primo valore della lista par, etc.

plt.plot(xdata,ydata,'bo',label='dati')     #aggiungo un titolo ai punti che rappresento
plt.plot(xfit,yfit,'r-',label='fit')        #disegno anche la curva con opzione '-' quindi solo la linea
plt.xlabel('ascisse')                       #titolo agli assi
plt.ylabel('ordinate')
plt.legend()                                #legenda
plt.grid()                                  #griglia
plt.show()
```

**Diffrazione di Fraunhofer**

La diffrazione di Fraunhofer corrisponde al caso in cui la luce diffratta da una o più fenditure (sul quale incide un fascio di raggi luminosi paralleli), è osservata a grande distanza dallo schermo stesso.

Nel caso di una singola fenditura di larghezza *a*

$$I = I_{max}(\frac{sin\beta}{\beta})^2$$     

$$\beta = \frac{a\pi}{\lambda}sin \theta$$

dove $\theta$ è l'angolo di diffrazione.

```python
import numpy as np
import matplotlib.pyplot as plt

L= 6.0 #distanza in metri dello schermo dalla fenditura
a=1.e-4 # apertura fenditura in m

l=650e-9 # lunghezza d'onda laser rosso
#l=532e-9 # lunghezza d'onda laser verde
#l=405e-9 # lunghezza d'onda laser viola

I_max=1000
```

Nel nostro esperimento con una fenditura sottile la distanza *L* è moto più grande della larghezza delle frange

$$\theta = arcsin\frac{x}{L}$$

$$x = L \ sin \ \theta$$

```python
# Define the x values
x = np.linspace(-0.5, 0.5, 1000)

beta=np.pi*a*np.sin(x/L)/l

# Compute the I values
I = I_max*(np.sin(beta)/beta)**2

# Create the plot
plt.figure(figsize=(10, 6))

# Plotting the sine wave
plt.plot(x, I, label=r'red laser', color='red')


plt.axis([-.5, .5, 0, 1000])
# Adding titles and labels
plt.title('Fraunhofer diffraction')

plt.xlabel('x [m]')
plt.ylabel('I value')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Display the plot
plt.show()
```

La posizione dei minimi è data dalla relazione:

$$a sin\theta = n \lambda$$

dove *n* è un numero intero

posizione sullo schermo $\frac{x}{L} = sin\theta$

$$ x = n\frac{\lambda L}{a}$$

```python
# Calcola le posizioni dei minimi, con i dati forniti
n = 10
for n in range(1,,1):
  x = n*l*L/a
  print('n = ', n,' x = ',round(x, 2), 'm')

```

```python
import numpy as np
import matplotlib.pyplot as plt

def intensity_pattern(x, slit_width, slit_separation, num_slits, wavelength, screen_distance):
    """Calculates the intensity of the interference and diffraction pattern."""
    k = 1 * np.pi / wavelength
    beta = (k * slit_width / 2) * (x / screen_distance)
    alpha = (k * slit_separation / 2) * (x / screen_distance)

    # Diffraction from a single slit
    single_slit_diffraction = (np.sin(beta) / beta)**2

    # Interference from multiple slits
    if num_slits == 1:
        multi_slit_interference = 1
    else:
        multi_slit_interference = (np.sin(num_slits * alpha) / np.sin(alpha))**2

    # Combined intensity
    intensity = single_slit_diffraction * multi_slit_interference
    return intensity

# Parameters (you can adjust these)
slit_width = 1e-5  # Width of each slit (in meters)
slit_separation = 5e-5  # Separation between slits (in meters)
num_slits = 2  # Number of slits
wavelength = 500e-9  # Wavelength of light (in meters)
screen_distance = 1  # Distance to the screen (in meters)
x_max = 0.05  # Maximum extent on the screen (in meters)
num_points = 1000  # Number of points for the calculation

# Generate the x-axis values on the screen
x = np.linspace(-x_max, x_max, num_points)

# Calculate the intensity pattern
intensity = intensity_pattern(x, slit_width, slit_separation, num_slits, wavelength, screen_distance)

# Normalize intensity for better visualization
intensity /= np.max(intensity)

# Plot the intensity pattern
plt.figure(figsize=(10, 6))
plt.plot(x, intensity, color='blue')
plt.xlabel('Posizione sullo schermo (m)')
plt.ylabel('Intensit√† normalizzata')
plt.title(f'Interferenza e Diffrazione con {num_slits} Fenditure')
plt.grid(True)
plt.show()
```

**Spiegazione del codice:**

1.  **Importa Librerie:** Importiamo `numpy` per i calcoli numerici e `matplotlib.pyplot` per la creazione di grafici.
2.  **Funzione `intensity_pattern`:**
    *   Questa funzione calcola l'intensit√† luminosa in un dato punto `x` sullo schermo.
    *   Prende come input i parametri della fenditura (larghezza, separazione, numero di fenditure), la lunghezza d'onda della luce e la distanza dallo schermo.
    *   Calcola i termini `beta` e `alpha` che dipendono dalla geometria del setup e dalla lunghezza d'onda.
    *   Calcola la diffrazione da una singola fenditura usando la formula `(sin(beta)/beta)^2`.
    *   Calcola l'interferenza da multiple fenditure usando la formula `(sin(N*alpha)/sin(alpha))^2`, dove N √® il numero di fenditure. Se c'√® una sola fenditura, l'interferenza √® semplicemente 1.
    *   L'intensit√† totale √® il prodotto della diffrazione da singola fenditura e dell'interferenza da multiple fenditure.
3.  **Parametri:** Definiamo i parametri per la simulazione (larghezza della fenditura, separazione, numero di fenditure, lunghezza d'onda, distanza dallo schermo). Puoi modificare questi valori per esplorare diversi scenari.
4.  **Genera Punti sullo Schermo:** Creiamo un array di punti `x` sullo schermo dove calcoleremo l'intensit√†.
5.  **Calcola Intensit√†:** Chiamiamo la funzione `intensity_pattern` per ottenere l'intensit√† in ogni punto `x`.
6.  **Normalizza Intensit√†:** Normalizziamo l'intensit√† dividendo per il valore massimo in modo che l'intensit√† massima sia 1. Questo rende il grafico pi√π facile da interpretare.
7.  **Traccia il Grafico:** Usiamo `matplotlib` per tracciare il grafico dell'intensit√† in funzione della posizione sullo schermo.

**Per aumentare il numero di fenditure**, basta cambiare il valore della variabile `num_slits` nel codice. Ad esempio, per 4 fenditure, imposta `num_slits = 4`. Esegui nuovamente la cella di codice per vedere il nuovo pattern.

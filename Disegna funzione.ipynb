```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values (angles in degrees)
x = np.linspace(0+0.0001, 180-0.0001, 1000)

# Define the x values (angles in radians)
#x = np.linspace(0, np.pi, 1000)

# Compute the sine values
#y = np.sin(x*np.pi/180)

# Compute the sine values
#y = np.sin(x)

# Compute the eta values
y = -np.log(np.tan(x*np.pi/360))
```

Sono stati creati due array x=np.linspace(0, 2*np.pi, 1000) e y=np.sin(x) di 1000 elementi.

```python
# Create the plot
plt.figure(figsize=(6, 6))

# Plotting the sine wave
plt.plot(x, y, label=r'$\eta = -\ln(\tan(\frac{\theta}{2}))$', color='red')
#plt.plot(x, y, label=r'$y = \sin(x)$', color='red')

# Adding titles and labels
#plt.title('Sine Function')
plt.title('eta Function')
plt.xlabel('Angle (degrees)')
#plt.xlabel('Angle (radians)')
#plt.ylabel('Sine value')
plt.ylabel('eta value')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Display the plot
plt.show()
```

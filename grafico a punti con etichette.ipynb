```python
import matplotlib.pyplot as plt

# Dati
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]
labels = ['A', 'B', 'C', 'D', 'E']

# Creazione del grafico a punti
plt.scatter(x, y, color='blue')

# Aggiunta delle etichette ai punti
for i, label in enumerate(labels):
    plt.annotate(label, (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Titolo e assi
plt.title("Grafico a Punti con Etichette")
plt.xlabel("Asse X")
plt.ylabel("Asse Y")

# Mostra il grafico
plt.show()
```

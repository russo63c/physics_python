**Calcolo di $\pi$ con il metodo di Leibnitz**



$$\pi \ = \ 4 \lim_{n \to \infty} \sum_{k=1}^n \frac{ (-1)^k}{2k+1} $$

```python
import numpy as np

sum=0.
for i in range(1000000):
    sum+=4*pow(-1,i)/(2*i+1)
    #print(sum)
    if np.abs(100*(np.pi-sum)/np.pi) < 0.01:
        print (i, sum, np.pi)
        break
```

**Calcolo di $\pi$ come integrale della funzione $f(x) = \frac{1}{1+x^2}$**

calcolo pi.JPG
$$ \pi = 2\int_{-1}^{1}\frac{1}{1+x^2} dx  $$

Nel codice Python seguente l'area della funzione è calcolata sommando le aree di rettangoli. L'altezza del rettangolo è il valore della funzione nel punto di mezzo della base del rettangolo.

```python
# Calcolo approssimato pigreco
# area rettangoli con punto di mezzo

import math


def my_fun(x):
    "definisci la funzione da integrare"
    y = 2/(1+pow(x,2))
    return y

def pi(n):
    a = -1
    b = 1

    a1 = a
    step =(b-a)/n
    b1 = a + step
    s = 0
    for i in range(n):
        xmed=a1+(b1-a1)/n
        s = s + my_fun(xmed)*(b1-a1)
        a1 = a1 + step
        b1 = b1 + step
    return s

print('math.pi = ', math.pi)
t = 6
for i in range(t):
    n = 100*pow(10,i)
    s = pi(n)
    print('     pi = ',s, ' n = ',n, ' diff % = ', 100*(math.pi-s)/math.pi)
```

Nel codice Python seguente l'area della funzione è calcolata sommando le aree di trapezi. Le basi del trapezio sono i due valori della funzione agli estremi dell'intervallo considerato

```python
# Calcolo approssimato pi
# area trapezio

import math
import numpy as np


def my_fun(x):
    y = 2/(1+pow(x,2))
    return y

def pi(n):
    a = -1
    b = 1
    h1=0.
    h2=0.
    s=0.
    a1 = a
    step =(b-a)/n
    b1 = a + step
    s = 0
    for i in range(n):
        h1=my_fun(a1)
        h2=my_fun(b1)
        s = s + (h1+h2)*step/2.
        a1 = a1 + step
        b1 = b1 + step
    return s

print('math.pi = ', math.pi)
t = 6
for i in range(t):
    n = 100*pow(10,i)
    s = pi(n)
    print('     pi = ',s, ' n = ',n, ' diff % = ', 100*(math.pi-s)/math.pi)
```
